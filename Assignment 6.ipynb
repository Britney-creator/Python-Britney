```python
import numpy as np
import matplotlib.pyplot as plt

def heat_transfer_1d(length, thermal_conductivity, area, delta_temperature):
    """
    Calculate heat transfer through a material.

    Parameters:
    - length: Thickness of the material (m)
    - thermal_conductivity: Thermal conductivity of the material (W/m K)
    - area: Cross-sectional area for heat transfer (m²)
    - delta_temperature: Temperature difference across the material (K)

    Returns:
    - heat_transfer_rate: Heat transfer rate (W)
    """

    heat_transfer_rate = (thermal_conductivity * area * delta_temperature) / length
    return heat_transfer_rate

def plot_temperature_profile(length, thermal_conductivity, area, delta_temperature, num_points=100):
    """
    Plot the temperature profile through a material.

    Parameters:
    - length: Thickness of the material (m)
    - thermal_conductivity: Thermal conductivity of the material (W/m K)
    - area: Cross-sectional area for heat transfer (m²)
    - delta_temperature: Temperature difference across the material (K)
    - num_points: Number of points for the temperature profile plot

    Returns:
    - None (displays the plot)
    """

    x = np.linspace(0, length, num_points)
    temperature_profile = delta_temperature * x / length

    plt.plot(x, temperature_profile)
    plt.xlabel('Distance (m)')
    plt.ylabel('Temperature (K)')
    plt.title('Temperature Profile')
    plt.grid(True)
    plt.show()

# Example usage
length_material = 0.1  # meters
thermal_conductivity_material = 50  # W/m K
cross_sectional_area = 0.01  # m²
temperature_difference = 100  # K

heat_rate = heat_transfer_1d(length_material, thermal_conductivity_material, cross_sectional_area, temperature_difference)
print(f"Heat Transfer Rate: {heat_rate} W")

plot_temperature_profile(length_material, thermal_conductivity_material, cross_sectional_area, temperature_difference)
```
